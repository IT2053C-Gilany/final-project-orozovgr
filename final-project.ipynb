# üèÜ Loading and File IO

- **Safe reading function** 
- **Exception handling** 


```python
from pathlib import Path
import pandas as pd


def load_csv(path: Path) -> pd.DataFrame:
  """
  Safely load a CSV file with helpful error messages.

  Parameters
  ----------
  path : Path
      The path to the CSV file.

  Returns
  -------
  pd.DataFrame
      The loaded DataFrame.
  """
  try:
    df = pd.read_csv(path)
  except FileNotFoundError as e:
    raise FileNotFoundError(
        f"‚ùå Could not find '{path.name}'. Make sure it is inside the data/ folder."
    ) from e
  except pd.errors.ParserError as e:
    raise ValueError(
        f"‚ùå Parsing failed for '{path.name}'. Check delimiter or malformed rows."
    ) from e
  else:
    print(f"‚úÖ Successfully loaded: {path.name}")
    return df


# Load the dataset
data_path = Path("data/survey_results_public.csv")
df = load_csv(data_path)

df.head()
```

```python
from pathlib import Path
import numpy as np
import pandas as pd

# Use the safe loader you already defined
data_path = Path("data/survey_results_public.csv")
df = load_csv(data_path)

# ---- Clean / convert important numeric columns ----
num_cols = ["CompTotal", "YearsCode", "WorkExp"]
for col in num_cols:
  df[col] = pd.to_numeric(df[col], errors="coerce")

# Keep only reasonable ranges for salary and experience
salary_min, salary_max = 1000, 500_000      # tuple for bounds
years_min, years_max = 0, 60

mask = (
    df["CompTotal"].between(salary_min, salary_max)
    & df["YearsCode"].between(years_min, years_max)
)

df_clean = df[mask].copy()
```

This step loads the survey dataset safely and prepares key numeric fields for analysis. I convert important columns like compensation and years of experience into numeric values, handling any invalid entries. Then, I apply reasonable bounds to remove unrealistic or extreme responses. This filtering ensures that the cleaned dataset (df_clean) is accurate, consistent, and ready for reliable visualization and statistical analysis.

# üèÜ Statistical Analysis

- **Distribution characterization** 
- **Group comparison (t-test or Welch)** 
- **Categorical association (chi-square)** 
- **Correlation (Pearson or Spearman)** 


```python
# 1. Distribution characterization

import numpy as np
import pandas as pd

salary = df_clean["CompTotal"].dropna()

dist_summary = pd.DataFrame({
    "Mean": [salary.mean()],
    "Median": [salary.median()],
    "Std Dev": [salary.std()],
    "Min": [salary.min()],
    "Max": [salary.max()],
    "Skewness": [salary.skew()],
    "Q1": [salary.quantile(0.25)],
    "Q3": [salary.quantile(0.75)],
    "IQR": [salary.quantile(0.75) - salary.quantile(0.25)]
})

dist_summary
```

This table summarizes the distribution of annual compensation among respondents.
The mean is higher than the median, indicating right-skew, which is common in salary data due to a smaller set of highly-paid outliers.
The IQR (Q3‚ÄìQ1) shows the middle 50% of salaries, and the large range from minimum to maximum reflects wide income variation across global developers.
This helps identify data spread and informs future modeling decisions.

```python
# 2. Group comparison ‚Äì AI Threat vs Salary (Welch‚Äôs t-test)

from scipy import stats

group_yes = df_clean.loc[df_clean["AIThreat"] == "Yes", "CompTotal"].dropna()
group_no = df_clean.loc[df_clean["AIThreat"] == "No", "CompTotal"].dropna()

t_stat, p_val = stats.ttest_ind(group_yes, group_no, equal_var=False)

t_stat, p_val
```

I compared annual compensation between developers who believe AI is a threat and those who do not.
Welch‚Äôs t-test is used because the groups have unequal variances and unequal sample sizes.
The resulting p-value determines whether the difference in means is statistically significant.
This analysis reveals whether perception of AI threat is associated with differences in salary.

```python
# 3. Categorical Association (Chi-Square Test)

# Create contingency table
cont_table = pd.crosstab(df_clean["AIThreat"], df_clean["DevType"])

# Chi-square test
chi2, p, dof, expected = stats.chi2_contingency(cont_table)

chi2, p, dof
```

A chi-square test was used to evaluate whether perception of AI as a threat varies across different developer roles.
The contingency table compares observed vs expected counts under independence.
A significant result implies that certain job roles are more (or less) likely to view AI as a threat.
This identifies meaningful relationships between profession and AI attitudes.

```python
# 4. Correlation Between Two Numeric Variables (Pearson)

x = df_clean["YearsCode"].dropna()
y = df_clean["CompTotal"].dropna()

corr, p_val = stats.pearsonr(x, y)

corr, p_val
```

Pearson correlation measures the linear relationship between years of coding experience and annual salary.
The correlation coefficient (r) shows both strength and direction, while the p-value indicates statistical significance.
A positive correlation would suggest that more experience is associated with higher compensation, although real-world salary growth may plateau after certain experience levels.
